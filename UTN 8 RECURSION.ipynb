```python
def calcular_factorial(n):   
    
    if n <= 1:
        return 1  
    
    else:
        return n * calcular_factorial(n - 1)

# --- PROGRAMA PRINCIPAL ---

def main():
    print("--- Calculadora de Factoriales ---")
    
    while True:
        entrada = input("Ingrese un número entero positivo (o 'salir' para terminar): ")
        
        if entrada.lower() == 'salir':
            break
            
        if not entrada.isdigit():
            print("Error: Por favor ingrese un número entero positivo.")
            continue
            
        numero_limite = int(entrada)
        
        print(f"\nCalculando factoriales desde 1 hasta {numero_limite}:")
        print("-" * 30)
        
        # Bucle para calcular y mostrar cada factorial
        for i in range(1, numero_limite + 1):
            resultado = calcular_factorial(i)
            print(f"El factorial de {i} ({i}!) es: {resultado}")
            
        print("-" * 30)
        print("¡Cálculo finalizado!\n")

if __name__ == "__main__":
    main()
```

```python
def fibonacci(n):
    
    # 1. Casos Base (los cimientos)
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    
    
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# --- PROGRAMA PRINCIPAL ---

def main():
    print("--- Generador de Serie de Fibonacci ---")
    
    while True:
        entrada = input("Ingrese hasta qué posición quiere ver la serie (o 'salir'): ")
        
        if entrada.lower() == 'salir':
            break
            
        if not entrada.isdigit():
            print("Error: Por favor ingrese un número entero positivo.")
            continue
            
        limite = int(entrada)
        
        print(f"\nCalculando la serie hasta la posición {limite}:")
        print("-" * 40)
        
        serie_completa = []
        

        for i in range(limite + 1):
            valor = fibonacci(i)
            serie_completa.append(valor)
            # Imprimimos cada paso para verlo claro
            print(f"Posición {i}: {valor}")
            
        print("-" * 40)
        print(f"Serie final: {serie_completa}\n")

if __name__ == "__main__":
    main()
```

```python
def calcular_potencia(base, exponente):

    if exponente == 0:
        return 1
    
    
    if exponente < 0:
        raise ValueError("Esta implementación recursiva solo soporta exponentes no negativos.")
        

    else:
        return base * calcular_potencia(base, exponente - 1)



def main():
    print("--- Calculadora de Potencia Recursiva (n^m) ---")
    
    while True:
        try:
            # Pedir la base
            entrada_base = input("\nIngrese la base (n) (o 'salir'): ")
            if entrada_base.lower() == 'salir':
                break
            
            base = float(entrada_base)
            
            # Pedir el exponente
            entrada_exp = input("Ingrese el exponente (m): ")
            
            # Validamos que el exponente sea un entero
            if not entrada_exp.isdigit():
                print("Error: El exponente debe ser un número entero positivo o cero.")
                continue

            exponente = int(entrada_exp)
            
            # Calcular la potencia
            resultado = calcular_potencia(base, exponente)
            
            # Mostrar el resultado
            print("-" * 35)
            print(f"Resultado: {base} elevado a {exponente} ({base}^{exponente}) es: {resultado}")
            print("-" * 35)
            
        except ValueError as e:
            print(f"Error de entrada: {e}")
        except Exception as e:
            print(f"Ocurrió un error inesperado: {e}")

if __name__ == "__main__":
    main()
```

```python
def decimal_a_binario(decimal):

    if decimal < 0:
        raise ValueError("El número debe ser un entero positivo.")
    

    if decimal == 0:
        return ""
    

    resto = decimal % 2
    
    # b. Obtener el cociente (la siguiente llamada recursiva).
    cociente = decimal // 2
    

    return decimal_a_binario(cociente) + str(resto)

# --- ALGORITMO GENERAL DE PRUEBA ---

def main():
    print("--- Conversor Decimal a Binario Recursivo ---")
    
    while True:
        entrada = input("\nIngrese un número entero positivo (o 'salir'): ")
        
        if entrada.lower() == 'salir':
            break
        
        # Validar si es un número válido (usamos try-except para ser robustos)
        try:
            numero = int(entrada)
            
            if numero == 0:
                print(f"Resultado binario: 0")
                continue
            
            resultado_binario = decimal_a_binario(numero)
            
            print("-" * 40)
            print(f"El número decimal {numero} se convierte a binario: '{resultado_binario}'")
            print(f"({numero} ÷ 2, restos leídos al revés)")
            print("-" * 40)
            
        except ValueError as e:
            # Captura errores como texto inválido o números negativos
            print(f"Error: {e}")
        except Exception as e:
            print(f"Ocurrió un error inesperado: {e}")

if __name__ == "__main__":
    main()
```

```python
def es_palindromo(palabra: str) -> bool:

    

    if len(palabra) <= 1:
        return True

    if palabra[0] == palabra[-1]:

        return es_palindromo(palabra[1:-1])
    else:

        return False

# --- ALGORITMO DE PRUEBA ---

def main():
    print("--- Verificador de Palíndromos Recursivo ---")
    
    # Lista de palabras para probar
    palabras_a_probar = [
        "reconocer",  # Palíndromo
        "neuquen",    # Palíndromo
        "anitalavalatina", # Palíndromo
        "python",     # No Palíndromo
        "madera",     # No Palíndromo
        "a",          # Palíndromo (Caso Base: 1 letra)
        ""            # Palíndromo (Caso Base: 0 letras)
    ]
    
    print("\nEjecutando pruebas automáticas:")
    for p in palabras_a_probar:
        resultado = es_palindromo(p)
        estado = "✅ SÍ es un palíndromo" if resultado else "❌ NO es un palíndromo"
        print(f"'{p}' -> {estado}")
        
    print("-" * 40)
    
    # Prueba interactiva
    while True:
        entrada = input("Ingrese una palabra (o 'salir'): ").strip().lower()
        if entrada == 'salir':
            break
        
        if not entrada:
            print("Por favor ingrese al menos un carácter.")
            continue
            
        resultado = es_palindromo(entrada)
        estado = "¡SÍ es un palíndromo!" if resultado else "¡NO es un palíndromo!"
        print(f"La palabra '{entrada}' {estado}\n")

if __name__ == "__main__":
    main()
```

```python
def suma_digitos(n: int) -> int:

    if n < 0:
        raise ValueError("La función solo acepta números enteros positivos.")
        

    if n < 10:
        return n
    

    ultimo_digito = n % 10
    

    resto_del_numero = n // 10
    

    return ultimo_digito + suma_digitos(resto_del_numero)

# --- ALGORITMO GENERAL DE PRUEBA ---

def main():
    print("--- Suma de Dígitos Recursiva ---")
    

    ejemplos = [1234, 9, 305, 5000, 987]
    print("\nEjecutando pruebas automáticas:")
    for num in ejemplos:
        try:
            resultado = suma_digitos(num)
            print(f"La suma de los dígitos de {num} es: {resultado}")
        except ValueError as e:
            print(f"Error con {num}: {e}")

    print("-" * 40)
    

    while True:
        entrada = input("Ingrese un número entero positivo (o 'salir'): ")
        
        if entrada.lower() == 'salir':
            break
            
        try:
            numero = int(entrada)
            resultado = suma_digitos(numero)
            
            print(f"La suma de los dígitos de {numero} es: {resultado}")
            
        except ValueError as e:
            print(f"Error de entrada: {e}. Asegúrese de ingresar un entero positivo.")
        except Exception as e:
            print(f"Ocurrió un error: {e}")

if __name__ == "__main__":
    main()
```

```python
def contar_bloques(n: int) -> int:
    

    if n < 1:
        return 0 
    

    if n == 1:
        return 1
    

    return n + contar_bloques(n - 1)

# --- ALGORITMO GENERAL DE PRUEBA ---

def main():
    print("--- Calculadora de Bloques de Pirámide Recursiva ---")
    
    # Ejemplos automáticos
    ejemplos = [1, 2, 4, 10]
    print("\nEjecutando pruebas automáticas:")
    for n in ejemplos:
        try:
            resultado = contar_bloques(n)
            print(f"Pirámide con {n} bloques en la base (1 + ... + {n}): Total = {resultado}")
        except Exception as e:
            print(f"Error con n={n}: {e}")

    print("-" * 50)
    
    # Prueba interactiva
    while True:
        entrada = input("Ingrese el número de bloques en la base (n) o 'salir': ")
        
        if entrada.lower() == 'salir':
            break
            
        try:
            numero = int(entrada)
            if numero < 1:
                print("Por favor, ingrese un entero positivo (mínimo 1).")
                continue
                
            resultado = contar_bloques(numero)
            
            print(f"\nLa pirámide de base {numero} requiere un total de {resultado} bloques.")
            
        except ValueError:
            print("Error: Ingrese un número entero válido.")
        except Exception as e:
            print(f"Ocurrió un error: {e}")

if __name__ == "__main__":
    main()
```

```python
def contar_digito(numero: int, digito: int) -> int:
    
    # 1. Manejo de errores
    if numero < 0 or not (0 <= digito <= 9):
        raise ValueError("El número debe ser positivo y el dígito debe estar entre 0 y 9.")
    
    # 2. Caso Base: Si el número es 0, la recursión termina.
    if numero == 0:
        return 0
    
    # 3. Caso Recursivo:
    
    # a. Obtener el último dígito
    ultimo_digito = numero % 10
    


    coincide = 1 if ultimo_digito == digito else 0
    
    # c. Llamada recursiva: Sumamos 
    # [1 si el dígito actual coincide] + [el resultado de contar en el resto del número]
    resto_del_numero = numero // 10
    
    return coincide + contar_digito(resto_del_numero, digito)

# --- ALGORITMO GENERAL DE PRUEBA ---

def main():
    print("--- Contador de Dígitos Recursivo ---")
    
    # Ejemplos automáticos
    ejemplos = [
        (12233421, 2, 3), # 3 veces el 2
        (5555, 5, 4),     # 4 veces el 5
        (123456, 7, 0),   # 0 veces el 7
        (101010, 0, 3)    # 3 veces el 0
    ]
    
    print("\nEjecutando pruebas automáticas:")
    for num, dig, esperado in ejemplos:
        try:
            resultado = contar_digito(num, dig)
            estado = "✅ CORRECTO" if resultado == esperado else f"❌ INCORRECTO (Esperado: {esperado})"
            print(f"Contar {dig} en {num}: Resultado: {resultado} ({estado})")
        except ValueError as e:
            print(f"Error con {num}, {dig}: {e}")

    print("-" * 50)
    
    # Prueba interactiva
    while True:
        try:
            entrada_num = input("Ingrese el número (o 'salir'): ")
            if entrada_num.lower() == 'salir':
                break
            
            entrada_dig = input("Ingrese el dígito a contar (0-9): ")
            
            numero = int(entrada_num)
            digito = int(entrada_dig)
            
            resultado = contar_digito(numero, digito)
            
            print(f"\nEl dígito {digito} aparece {resultado} veces en {numero}.")
            
        except ValueError:
            print("Error: Ingrese números enteros válidos.")
        except Exception as e:
            print(f"Ocurrió un error: {e}")

if __name__ == "__main__":
    main()
```
